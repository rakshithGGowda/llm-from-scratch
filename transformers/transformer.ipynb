```python
import string

allowed_chars = set(string.ascii_letters + string.digits + string.punctuation + '\n\t' + ' ')

cleaned_text = ''.join([ch for ch in text if ch in allowed_chars])

# verify the new vocab
chars_cleaned = sorted(list(set(cleaned_text)))
vocab_size_cleaned = len(chars_cleaned)
print(vocab_size_cleaned)
print(chars_cleaned)
```

```python
#Tokerizer:

stoi = {ch:i for i,ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}


encode = lambda s: [stoi[ch] for ch in s]
decode = lambda i: ''.join([itos[item] for item in i])

print(stoi['r'])
print(itos[stoi['r']])

print(encode("hello world"))
print(decode(encode("hello world")))
```
